### Preparación de variables para las hipótesis planteadas
- La dirección del choque influye en tasa de mortalidad del los accidentes
- A más velocidad maxima de la via más accidentes con heridos graves o fatales
- La distribución de accidentes a lo largo del año sigue un patron repetitivo para todos los años
- Los accidentes fatales se concentran en ciertas localidades
- Los motoristas y peatones representan un porcentaje alto de las víctimas fatales respecto a su participación en el tráfico

Eliminaremos las variables redundantes y las que no aporten valor para analizar las hipótesis mencionadas

````python
unw_cols_collision = [
    'date',
    'time',
    'collision_year',
    'collision_ref_no',
    'location_easting_osgr',
    'location_northing_osgr',
    'local_authority_district',
    'local_authority_highway_current',
    #'first_road_number',
    'second_road_number',
    'pedestrian_crossing_human_control_historic',
    'pedestrian_crossing_physical_facilities_historic',
    'carriageway_hazards_historic',
    'junction_detail_historic',
    'collision_adjusted_severity_serious',
    'collision_adjusted_severity_slight',
    'lsoa_of_accident_location',
    'enhanced_severity_collision',
    'collision_injury_based',
    'pedestrian_crossing',
    'special_conditions_at_site',
    'carriageway_hazards',
    'did_police_officer_attend_scene_of_accident',
    'trunk_road_flag',
]
print(collision_cln.shape)
collision_cln = collision_cln.drop(columns=unw_cols_collision)
print(collision_cln.shape)
collision_cln.head()
````

````python
unw_cols_casualty = [
    'casualty_reference',
    'vehicle_reference',
    'lsoa_of_casualty',
    'casualty_distance_banding',
    'casualty_injury_based',
    'collision_year',
    'collision_ref_no',
    'casualty_adjusted_severity_serious',
    'casualty_adjusted_severity_slight',
    'pedestrian_location', 'pedestrian_movement',
    'car_passenger',
    'bus_or_coach_passenger',
    'pedestrian_road_maintenance_worker',
    'casualty_imd_decile'
]
print(casualty_cln.shape)
casualty_cln  = casualty_cln.drop(columns=unw_cols_casualty)
print(casualty_cln.shape)
casualty_cln.head()
````

````python
unw_cols_vehicle = [
    'collision_ref_no',
    'vehicle_reference',
    'generic_make_model',
    'driver_imd_decile',
    'lsoa_of_driver',
    'driver_distance_banding',
    'towing_and_articulation',
    'vehicle_left_hand_drive',
    'journey_purpose_of_driver_historic',
    'vehicle_manoeuvre_historic',
    'collision_year',
    'vehicle_direction_from',
    'vehicle_direction_to',
    'vehicle_location_restricted_lane_historic',
    'vehicle_location_restricted_lane',
    'junction_location',
    'skidding_and_overturning',
    'hit_object_in_carriageway',
    #'vehicle_leaving_carriageway',
    #'hit_object_off_carriageway',
    'journey_purpose_of_driver',
    'engine_capacity_cc',
    'propulsion_code',
    'age_of_vehicle',
    'escooter_flag'
]
print(vehicle_cln.shape)
vehicle_cln = vehicle_cln.drop(columns=unw_cols_vehicle, errors='ignore')
print(vehicle_cln.shape)
vehicle_cln.head()
````

### Exportación
Exportamos los datasets reducidos y limpios que usaremos para analizar las hipótesis. Filtraremos mas cuidadosamente las variables cuando estemos analizando cada  hipotesis por separado.

````python
collision_cln.to_parquet('../data/processed/uk_collision_cln.parquet', index=False)
casualty_cln.to_parquet('../data/processed/uk_casualty_cln.parquet', index=False)
vehicle_cln.to_parquet('../data/processed/uk_vehicle_cln.parquet', index=False)
````
